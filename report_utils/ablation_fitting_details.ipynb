```python
import sys
from pathlib import Path

try:                                   # regular .py execution
    repo_root = Path(__file__).resolve().parent.parent
except NameError:                      # Jupyter / IPython
    repo_root = Path().resolve().parent

sys.path.append(str(repo_root))
print("Added to sys.path →", repo_root)
```

```python
fitting_dir = Path("../results/fitting_jailbreaks/overall_results")
table_dir = Path("../report/tables/fitting_jailbreaks/")
table_dir.mkdir(parents=True, exist_ok=True)
```

```python
import pandas as pd

csv_files = list(fitting_dir.rglob("*.csv"))
dfs = []
for f in csv_files:
    print(f)
    model = f.stem.split("_")[1]
    df = pd.read_csv(f)
    df = df.iloc[[0]]
    df['model'] = model
    dfs.append(df)

fitting_results = pd.concat(dfs, ignore_index=True)
```

```python
fitting_results
```

```python
fitting_results = fitting_results[["model", "layer", "baseline_harmless_refusal_rate", "harmful_refusal_rate", "harmless_refusal_rate", "kl_divergence", "overall_score"]]
```

```python
fitting_results
```

```python
fitting_results.columns = [
    col.replace("_", " ").title() for col in fitting_results.columns
]
```

```python
fitting_results
```

```python
table = fitting_results.to_latex()

with open(table_dir / "fitting_details.tex", "w", encoding="utf-8") as f:
    f.write(table)
```
