# Fitting

```python
fitting_dir = Path("../results/fitting_jailbreaks")
table_dir = Path("../report/tables/")
table_dir.mkdir(parents=True, exist_ok=True)
```

```python
import pandas as pd

csv_files = list(fitting_dir.rglob("*fit_times*.csv"))
dfs = []
for f in csv_files:
    print(f)
    run_id = f.stem  # file name without extension
    df = pd.read_csv(f)
    df['run_id'] = run_id
    dfs.append(df)

fitting_results = pd.concat(dfs, ignore_index=True)
```

```python
fitting_results = fitting_results.rename(columns={
    "model_path": "model",
    "method_name": "method",
    "time": "fitting time"
})
```

```python
model_map = {
    "Llama-3.1-8B-Instruct": "Llama-3",
    "Qwen2.5-7B-Instruct": "Qwen2.5",
    "gemma-2-9b-it": "Gemma-2",
    "phi-4": "Phi-4",
    "Qwen/Qwen2.5-7B-Instruct": "Qwen2.5",
    "meta-llama/Llama-3.1-8B-Instruct": "Llama-3",
    "google/gemma-2-9b-it": "Gemma-2",
    "microsoft/phi-4": "Phi-4",
}
fitting_results["model"] = fitting_results["model"].replace(model_map)

method_map = {
    "diffinmeans": "Single Refusal Direction Ablation",
    "prefixinjection-3": "Prefix Injection",
    "baseline": "Baseline",
    "outputaware-temperature": "Output-Aware Temperature Sampling",
    "hiddenlayerinjection": "Hidden Layer Injection",

}
fitting_results["method"] = fitting_results["method"].replace(method_map)
```

```python
fitting_results
```

```python
metrics = ["fitting time"]
```

```python
fitting_results = fitting_results.groupby(['method', 'model', 'run_id']).agg({
    **{metric: 'mean' for metric in metrics},
})
```

```python
from scipy.stats import norm
import numpy as np

alpha = 0.05
z = norm.ppf(1 - alpha/2)

def ci_error(series):
    return z * series.std(ddof=1) / np.sqrt(series.count())
ci_error.__name__ = 'ci_error'


fitting_grouped_df = fitting_results.groupby(['method', 'model']).agg({
    **{metric: ['mean', 'std', 'count', ci_error] for metric in metrics},
})

fitting_method_df = fitting_results.groupby(['method']).agg({
    **{metric: ['mean', 'std', 'count', ci_error] for metric in metrics},
})
```

```python
def mean_ci_table(
    df: pd.DataFrame,
    *,
    precision: int = 1,
    percent_metrics: set[str] | None = None,
    metric_map: dict[str, dict[str, str]] | None = None
) -> pd.DataFrame:
    percent_metrics = set(percent_metrics or [])
    out = pd.DataFrame(index=df.index)

    for metric in df.columns.get_level_values(0).unique():
        if ('mean' not in df[metric]) or ('ci_error' not in df[metric]):
            continue                                # skip incomplete metrics

        mean = df[(metric, 'mean')].copy()
        ci   = df[(metric, 'ci_error')].copy()

        if metric in percent_metrics:
            mean *= 100
            ci   *= 100

        out[metric] = (
            mean.round(precision).map(lambda x: f"{x:.{precision}f}")
            + " ± "
            + ci.round(precision).map(lambda x: f"{x:.{precision}f}")
        )
    if metric_map is not None:
        rename_dict = {
            metric: metric_map[metric]["name"] 
            for metric in out.columns 
            if metric in metric_map
        }
        out = out.rename(columns=rename_dict)

    out = out.reindex(sorted(out.columns), axis=1)
    return out
```

```python
table = mean_ci_table(fitting_grouped_df)

latex_table = table.to_latex(escape=False)
```

```python
with open(f"{table_dir}/fitting_results.tex", "w", encoding="utf-8") as f:
    f.write(latex_table)
```
